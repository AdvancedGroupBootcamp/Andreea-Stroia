# Description of the business
The Professional Maids is a business which offers professional cleaning services for apartments in Paris.

```python

class Maid():
    #Constructor
        
    def __init__(self):
        self.name=None
        self.adress=None
        self.region=None
        self.telephone=None
        self.years_of_experience=0
        self.charge_per_hour=0
        self.number_of_worked_hous_per_day=0
        self.daily_income=0
        self.discount_transportation=0
        self.extra_pay=0
        self.day_off=0
        self.vacuum_cleaner=None
        self.uniform=None
    
    #Methods
       
    def needs_vacuum_cleaner(self, x): #This method is supposed to check if the maid needs a vacum cleaner for their job
        if x == 'Yes':
            return 'Maid requests vacuum cleaner'
        else:
            return 'Maid does not request vacuum cleaner'
            
    def needs_uniform(self, x): #This method is supposed to check if the maid needs a unofifrom for their job
        if x == 'Yes':
            return 'Maid requests uniform'
        else:
            return 'Maid does not require uniform'
    
    def number_of_worked_hours(self): # This method is supposed to count the number of hours worker per day
        self.number_of_worked_hous_per_day +=1
        
    def Discounted_transport(self): # This method is supposed to check if the maid needs a discount transportation. If she lives outside Paris the company will make her a discount of 10 euros. If she lives in Paris the company will pay her the price of 2 tickets.
        if "District" not in self.region:
            self.discount_transportation = 10
        else:
            self.discount_transportation = 3.8
            
    def Day_off(self): # This method is supposed to count if the maid took a day off  
        self.day_off +=1
        self.daily_income=0
        self.discount_transportation =0
    
    def extra_services(self, extra_service): # This method is supposed to add an extra amount of money for each extra service the maid has done while cleaning the apartment
        lists = ['Ironing', 'Gardening', 'Cooking', 'Grocery shopping', 'Fur Cleanup'] 
        if extra_service in lists:
            if extra_service == 'Ironing':
                self.extra_pay += 4
            elif extra_service == 'Gardening':
                self.extra_pay += 10
            elif extra_service == 'Cooking':
                self.extra_pay += 7
            elif extra_service == 'Grocery shopping':
                self.extra_pay += 6
            else:
                self.extra_pay += 8
        else:
            print('Option not availabe')

    
    def data_extracting(self, form):
        self.name=form.split(' - ')[0]
        self.adress=form.split(' - ')[2]
        self.telephone=form.split(' - ')[3]
        self.years_of_experience=form.split(' - ')[4]
        self.vacuum_cleaner=self.needs_vacuum_cleaner(form.split(' - ')[-1])
        self.uniform=self.needs_uniform(form.split(' - ')[-2])
    
    def data_featuring(self):
        self.region=self.adress.split(', ')[-1]
        if int(self.years_of_experience.split(' ')[0]) <=1:
            self.level='Entry'
            self.charge_per_hour=15
        elif int(self.years_of_experience.split(' ')[0]) >1 and int(self.years_of_experience.split(' ')[0])<=3:
            self.level='Mid'
            self.charge_per_hour=19
        else:
            self.level='Senior'
            self.charge_per_hour=25
    
    def get_data_cleaned(self):
        return{
            'Name':self.name,
            'Adress':self.adress,
            'Telephone':self.telephone,
            'Level':self.level,
            'Vacuum cleaner': self.vacuum_cleaner,
            'Uniform':self.uniform,
            'Daily Income':self.number_of_worked_hous_per_day*self.charge_per_hour,
            'Extra Pay' :self.extra_pay,
            'Total Income': self.extra_pay + (self.number_of_worked_hous_per_day*self.charge_per_hour),
            'Discounted Transportation':self.discount_transportation      
            
        }
    
  
```
